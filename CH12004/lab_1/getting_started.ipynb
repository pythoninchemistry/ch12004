# Getting started

Before we dive headfirst into the fundamentals of programming in Python, let's take a whistle-stop tour of **running code** in a **Jupyter notebook**.

First off, open up a new Jupyter notebook via **Noteable**, as described previously in [lab 0](../lab_0/using_noteable.ipynb). You should be greeted with something like:

![blank_notebook](images/blank_notebook.png)

In general it is good practice to give your notebooks a somewhat more descriptive title than "Untitled", so feel free to **rename** your new notebook at this stage by clicking on "Untitled" and supplying a new name:

![rename_notebook](images/rename_notebook.png)

Jupyter notebooks are made up of **cells**; in your new notebook, there will only be **one** cell at first: the grey box with `In [ ]` to the left of it. We can start writing some code in a cell by clicking inside of it so that the **blue border** on the left turns **green**:

![green_border](images/green_border.png)

Now just start typing! 

![start_typing](images/start_typing.png)

Here our code is pretty easy to interpret even if you've never learnt any Python before: `2+2`. At the moment, we have **written** this code, but we have not **run** it, so the calculation has not actually been performed.

To **run** a cell in your Jupyter notebook, make sure it is currently selected (it should have the **green border** to the left) and press <kbd>Shift</kbd> + <kbd>Enter</kbd>:

![run_cell](images/run_cell.png)

Sure enough we now have an **output**, here the result of our simple calculation of `2+2`. Notice the numbers in square brackets to the left of the cell itself: `In [1]` and `Out[1]`. These represent the **order** in which our cells have been **run**. We can see this by running some more code (notice that Jupyter has added a **new cell** beneath the first one):

![second_cell](images/second_cell.png)

To reproduce the above, just type out the code in the second cell `5-2` and press <kbd>Shift</kbd> + <kbd>Enter</kbd> again. You should see that the second cell has `In [2]` and `Out[2]` to the left of it, indicating that this cell was run **after** the previous one. If we select the **first** cell again and run it for a **second time**:

![order](images/order.png)

We see that the numbers to the left reflect this, showing that this cell was the **third** to be run overall: `In [3]` and `Out[3]`.

In this book, we will represent **code cells** like this:

```python
2+2
```

If we are interested in not only the code itself, but also what happens when it **runs**, then we will show both the **code** and the **output** like this:

````python
2+2
````

This is equivalent to:

![run_cell](images/run_cell.png)

Before we crack on with the rest of the course, try running the following code in a new cell:

```python
1 + '1'
```

You should find that you get an **error message**:

![error](images/error.png)

Python, and for that matter all programming languages, use error messages to highlight **problems** in your code. In this case, don't worry about what the message means - we will come back to **errors** in much more detail later on - just be aware that this is what they look like and that if you see one, you don't need to panic!
